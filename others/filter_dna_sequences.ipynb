```python
import pandas as pd

df_data = pd.read_csv("unique_sequences.csv")
```

```python
length_sequences = [len(sequence) for sequence in df_data['sequence']]

df_data['length'] = length_sequences
df_data
```

```python
#todas las secuencias menores a 5 se van de manera inmediata -> no tiene sentido ya que ni si quiera es un péptido
filter_by_length = []

for i in range(len(df_data)):
    if df_data['length'][i] >=5:
        filter_by_length.append(0)
    else:
        filter_by_length.append(1)

df_data['filter_by_length'] = filter_by_length
df_data
        
```

```python
#checking possibles DNA sequences
matrix_data = []
for i in range(len(df_data)):
    sequence = df_data['sequence'][i]
    length_values = df_data['length'][i]
    counts_a = round(sequence.count('A')*100/length_values,2)
    counts_t = round(sequence.count('T')*100/length_values,2)
    counts_g = round(sequence.count('G')*100/length_values,2)
    counts_c = round(sequence.count('C')*100/length_values,2)

    others = 100 - counts_a - counts_c - counts_g - counts_t

    row = [df_data['id'][i], counts_a, counts_c, counts_g, counts_t, others]
    matrix_data.append(row)

df_counts = pd.DataFrame(matrix_data, columns=['id', 'counts_a', 'counts_c', 'counts_g', 'counts_t', 'others'])
df_counts
```

```python
possibles_dna = []

for i in range(len(df_counts)):
    if df_counts['counts_a'][i] == 100 or df_counts['counts_c'][i] == 100 or df_counts['counts_t'][i] == 100 or df_counts['counts_g'][i] == 100:
        possibles_dna.append(1)
    else:
        possibles_dna.append(0)

df_counts['1_possible_dna'] = possibles_dna
df_counts
```

```python
possibles_dna = []

for i in range(len(df_counts)):
    if df_counts['others'][i] == 0:
        possibles_dna.append(1)
    else:
        possibles_dna.append(0)

df_counts['2_possible_dna'] = possibles_dna
df_counts
```

```python
df_merge = df_data.merge(df_counts, left_on="id", right_on="id")
df_merge_filter = df_merge.loc[(df_merge['filter_by_length'] == 0) & (df_merge['1_possible_dna'] == 0) & (df_merge['2_possible_dna'] == 0)]
df_possible_error = df_merge.loc[(df_merge['filter_by_length'] == 1) | (df_merge['1_possible_dna'] == 1) | (df_merge['2_possible_dna'] == 1)]
```
