```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("../../results_demo/T50/FFT/Group_0/fft-Group_0-T50.csv")
dataset
```

```python
#categorizamos por respuesta
q1_dist = np.quantile(dataset['target'], .25)
q2_dist = np.quantile(dataset['target'], .50)
q3_dist = np.quantile(dataset['target'], .75)

categories_target = []

for index in dataset.index:
    target = dataset['target'][index]
    if target > q3_dist:
        categories_target.append(0)
    elif target <= q3_dist and target > q2_dist:
        categories_target.append(1)
    elif target <= q2_dist and target > q1_dist:
        categories_target.append(2)
    else:
        categories_target.append(3)

dataset['target_cat'] = categories_target
```

```python
dataset
```

```python
#generaremos elementos promedios por cada categoria
def create_average_curve(data_values, category):

    data_values_filter = data_values.loc[data_values['target_cat'] == category]

    average_curve = []
    q1_curve = []
    q3_curve = []

    for column in data_values_filter.columns:
        if "p_" in column:
            average_curve.append(np.mean(data_values_filter[column]))
            q1_curve.append(np.quantile(data_values_filter[column], .25))
            q3_curve.append(np.quantile(data_values_filter[column], .75))
    
    return average_curve, q1_curve, q3_curve
```

```python
average_curve_0, q1_curve_0, q3_curve_0 = create_average_curve(dataset, 0)
average_curve_1, q1_curve_1, q3_curve_1 = create_average_curve(dataset, 1)
average_curve_2, q1_curve_2, q3_curve_2 = create_average_curve(dataset, 2)
average_curve_3, q1_curve_3, q3_curve_3 = create_average_curve(dataset, 3)
```

```python
#definimos el eje X
stop_value = (len(dataset.columns)-3)*2#se quitan las tres columnas agregadas y se multiplica por 2
T = 1.0 / float(stop_value)
xf = np.linspace(0.0, 1.0 / (2.0 * T), stop_value // 2)
```

```python
#hacemos el plot
df_full0 = pd.DataFrame()
df_full0['average cat 0'] = average_curve_0
df_full0['.25% cat 0'] = q1_curve_0
df_full0['.75% cat 0'] = q3_curve_0

df_full1 = pd.DataFrame()
df_full1['average cat 1'] = average_curve_1
df_full1['.25% cat 1'] = q1_curve_1
df_full1['.75% cat 1'] = q3_curve_1

df_full2 = pd.DataFrame()
df_full2['average cat 2'] = average_curve_2
df_full2['.25% cat 2'] = q1_curve_2
df_full2['.75% cat 2'] = q3_curve_2

df_full3 = pd.DataFrame()
df_full3['average cat 3'] = average_curve_3
df_full3['.25% cat 3'] = q1_curve_3
df_full3['.75% cat 3'] = q3_curve_3
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


fig, axes = plt.subplots(2, 2, figsize=(15,7))

#ploteamos la visual de la categoria 0
sns.lineplot(ax=axes[0,0], data=df_full0)
axes[0, 0].set_title("Categoria 0")

#ploteamos la visual de la categoria 1
sns.lineplot(ax=axes[0,1], data=df_full1)
axes[0, 1].set_title("Categoria 1")

#ploteamos la visual de la categoria 2
sns.lineplot(ax=axes[1,0], data=df_full2)
axes[1, 0].set_title("Categoria 2")

#ploteamos la visual de la categoria 3
sns.lineplot(ax=axes[1,1], data=df_full3)
axes[1, 1].set_title("Categoria 3")

plt.show()
```
